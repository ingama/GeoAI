```python
import matplotlib.pyplot as plt 
import seaborn as sb
import pandas as pd
import scipy

import scipy.stats as stats
from scipy.stats import kurtosis

from sklearn.linear_model import LinearRegression
from numpy.random import lognormal
from scipy.stats import kstest


#select one station
idx_slice = pd.IndexSlice[:, station_labels[3], :]
cdf = cdf_stations.loc[idx_slice, :]
display(cdf)



def temperature_scatterplot(station_labels, data, msf):
        
        months = data.columns
        msf = msf
        
        fig, our_axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 15))
        monthly_axes = [ax for ax_row in our_axes for ax in ax_row]
        
        fig.subplots_adjust( hspace=0.4) 
        station_index = data.index[0][0:2]
        fig.suptitle(f'ID {station_index[1]}\n' + msf.loc[station_index,:].to_string(index = True))
        
        for i, month in enumerate(months):
            data_m = data.iloc[:,i].reset_index(level = 2, drop = False).reset_index(drop=True)

            s = sb.scatterplot(data = data_m,  x="YEAR", y=data_m.columns[1], ax = monthly_axes[i])
            s.set_xlabel('Year')
            s.set_ylabel('Temperature')
            s.set_title('Month ' + data_m.columns[1])
            
            #Linear Regression
            data_m_drop_na = data_m.dropna()
            x = np.array(data_m_drop_na.YEAR).reshape((-1, 1))
            y = np.array(data_m_drop_na[data_m_drop_na.columns[1]])
            model = LinearRegression().fit(x, y)
            r_sq = model.score(x, y)
            y_pred = model.predict(x)
            monthly_axes[i].plot(x, y_pred, 'r')
            s.set_xlabel(f'Year, coefficient of determination = {r_sq:5.2f}')

        return ()

def temperature_density(station_labels, data, msf):
        
        months = data.columns
        msf = msf
        
        fig, our_axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 15))
        monthly_axes = [ax for ax_row in our_axes for ax in ax_row]
        
        fig.subplots_adjust( hspace=0.5) 
        station_index = data.index[0][0:2]
        fig.suptitle(f'ID {station_index[1]}\n' + msf.loc[station_index,:].to_string(index = True))
        
        for i, month in enumerate(months):
            data_m = data.iloc[:,i].reset_index(drop=True).dropna()
            
            #density
            s = sb.kdeplot(data = data_m, ax = monthly_axes[i])
            
            #Shapiro_Wilk
            shap_s, shap_w = scipy.stats.shapiro(data_m)
            
            #Kolmogorov-Smirnov test
            kolm_s, kolm_w = kstest(data_m, 'norm')
            
            s.set_xlabel(f'Shapiro_Wilk {shap_s:6.3f}, {shap_w:6.3f}, \n Kolmogorov-Smirnov {kolm_s:6.3f}')
            s.set_title('Month ' + months[i])
            
        return ()

def skewness_kurtosis_histplot(station_labels, data, msf):
        
        months = data.columns
        msf = msf
        
        fig, our_axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 15))
        monthly_axes = [ax for ax_row in our_axes for ax in ax_row]
        
        fig.subplots_adjust( hspace=0.3)
        station_index = data.index[0][0:2]

        fig.suptitle(f'ID {station_index[1]}\n' + msf.loc[station_index,:].to_string(index = True))
        
        for i, month in enumerate(months):
            data_m = data.iloc[:,i].dropna()
            s = sb.histplot(data = data_m,  kde=True, ax = monthly_axes[i])
            
            #kurtosis
            kur = data_m.kurtosis()
            #skewness
            skw = data_m.skew()
            
            s.set_xlabel(f'Temperature, kurtosis = {kur:5.2f}, skewness = {skw:4.2f}')
            s.set_ylabel('Count')
            s.set_title('Month ' + months[i])
        return ()

#Station statistics
   
temperature_scatterplot(station_labels,cdf, msf)
temperature_density(station_labels,cdf, msf)
skewness_kurtosis_histplot(station_labels,cdf, msf)
plt.show()

```
