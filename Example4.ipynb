```python
months = cdf.columns

num_rows, num_columns = 4, 3
fig, our_axes = plt.subplots(nrows=4, ncols=3, figsize=(15, 15))
monthly_axes = [ax for ax_row in our_axes for ax in ax_row]

for i, month in enumerate(months):
    sb.histplot(cdf.iloc[:,i], kde=True, ax = monthly_axes[i])
plt.show()   
```

```python
max_idx = cdf.idxmax(axis="rows")
max_table = cdf.loc[max_idx].assign(MMAX = cdf.max(axis = 1))

display(max_table)

cdf_test = cdf.droplevel([0, 1], axis=0) 
cdf_test = cdf_test.reset_index(drop = True) 

plt.figure(figsize=(8,6))
plt.title("Correlation between all columns")
sb.heatmap(data= cdf_test.corr(), cmap="rocket", annot=True)
plt.show()

cdf_test = cdf_test.map(lambda x: np.nan if x>=70 else x)

plt.figure(figsize=(8,6))
plt.title("Correlation between all columns without outliers")
sb.heatmap(data= cdf_test.corr(), cmap="rocket", annot=True)
plt.show()

sb.histplot(cdf_test.M2, kde=True)
plt.show()
```
