```python
import geopandas
from geodatasets import get_path
from shapely.geometry import Point
import matplotlib.pyplot as plt
import pathlib

country_labels = (
        'CA', 
        'US', 
        'MX')

print("\nSELECTED COUNTRIES STATIONS LATITUDE AND LONGITUDE")
idx_slice = pd.IndexSlice[country_labels,:]
countries_lat_lon = msf.loc[idx_slice, ['LONGITUDE', 'LATITUDE', 'NAME']].droplevel([0], axis=0)
display(countries_lat_lon)

station_labels = (
        'BR027794310',
        'BRM00082994',
        'RSM00026063',
        'RSM00027612',
        'WAXLT711578',
        'USC00305804',
        'USW00023234')

#0#BR027794310  -3.7000  -40.3500   83.0 SOBRAL
#1#BRM00082994  -9.6600  -35.7000   64.5 MACEIO 
#2#RSM00026063  59.9667   30.3000    3.0 ST_PETERSBURG 
#3#RSM00027612  55.8331   37.6167  156.0 MOSCOW 
#4#WAXLT711578 -22.9400   14.4400 9999.0 WALFISCH_BAY  
#5#USC00305804  40.7333  -73.9333    3.0 NEW_YORK_LAUREL_HILL
#6#USW00023234  37.6197 -122.3647    2.4 SAN_FRANCISCO_INTL_AP

print("\nSELECTED STATIONS LATITUDE AND LONGITUDE")
idx_slice = pd.IndexSlice[:,station_labels]
stations_lat_lon = msf.loc[idx_slice, ['LONGITUDE', 'LATITUDE', 'NAME']]
display(stations_lat_lon)

#worldmap file load
home_directory = "D:/" +  "/ClimateData/Data/Maps/ne_10m_admin_0_countries/"
file_name = "ne_10m_admin_0_countries.shp"
location = pathlib.Path(home_directory)
path = location / file_name

world = geopandas.read_file(path)


print("\nGEOPANDAS OBJECT - WORLD - COLUMNS\n")
print("\n".join(str(world.columns.values).split('\n')))

print("\nGEOPANDAS OBJECT - WORLD - COUNTRY LABELS\n")
print("\n".join(str(world['ISO_A2']).split('\n')))

print("\n".join(str(world['REGION_UN']).split('\n')))

fig, ax = plt.subplots(figsize=(12, 12))
world.plot(ax = ax,color='steelblue')
display(world)

#plot countries stations
geometry = [Point(xy) for xy in zip(countries_lat_lon['LONGITUDE'], countries_lat_lon['LATITUDE'])]
geo_station = geopandas.GeoDataFrame(countries_lat_lon, geometry=geometry)
geo_station.plot(ax=ax, marker='o', color='red', markersize=15);

#plot selected stations
geometry = [Point(xy) for xy in zip(stations_lat_lon['LONGITUDE'], stations_lat_lon['LATITUDE'])]
geo_station = geopandas.GeoDataFrame(stations_lat_lon, geometry=geometry)
geo_station.plot(ax=ax, marker='o', color='black', markersize=15);

ax.title.set_text("WORLD MAP")
label = ["COUNTRIES STATIONS","SELECTED STATIONS"]
ax.legend(loc='best', fontsize='small', labels=label)

k=0;
for i,j in zip(stations_lat_lon['LONGITUDE'], stations_lat_lon['LATITUDE']):
    shift = 3 # adds a little shift to annotation 
    ax.annotate(stations_lat_lon.iloc[k]['NAME'],  xy=(i + shift, j + shift), color='black', fontsize="small")
    k += 1

ax.set_xlabel("LONGITUDE")
ax.set_ylabel("LATITUDE")


plt.show()
```
