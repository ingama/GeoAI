```python
#---------MAP FUNCTION--------------------------------------------------------------------------
#Lambda functions are similar to user-defined 
#functions but without a name - anonymous functions

print('MAP FUNCTION TO ALL THE ELEMENTS - REPLACE NAN VALUES WITH -9999 AND VICE VERSA.\n\
       USE PANDAS VERSION GREATER THAN 2.1.0!')
cdf_test = cdf.map(lambda x: -9999 if np.isnan(x) else x)
display(cdf_test)

cdf_test = cdf_test.map(lambda x: np.nan if x==-9999 else x)
display(cdf_test)
```

```python
#------DROP INDEXES----------------------------------------------------------------------------------

print('DROP LEVEL 0 (COUNTRY) and 1 (ID)')
cdf_test = cdf.droplevel([0, 1], axis=0)             #axis = 0 has three level of MultiIndex, at least one level must be left
display(cdf_test)

print('RESET INDEX LEVEL 2 (YEAR) OF MULTIINDEX')
cdf_test = cdf.reset_index(level = 2, drop = False)
display(cdf_test)

print('RESET MULTIINDEX')
cdf_test = cdf.reset_index()                          
display(cdf_test)

print('REMOVE MULTIINDEX COLUMNS')
cdf_test = cdf.reset_index(drop = True)               
display(cdf_test)
```
